# Case Studies

Use this notebook for studying single data points. Feel free to change `DATA_PT` to the path of the data you want to analyze, and also the `OUT_PATH` to the output path of the csv that is generated from the data.

```python
import json
import csv

# Feel free to tweak around
DATA_PT = './Yizhen_sample.json'
OUT_PATH = 'out.csv'

with open(DATA_PT) as json_file:
    data = json.load(json_file)["Data"]

with open(OUT_PATH, 'w') as out:
    csv_writer = csv.writer(out)

    header = data[0].keys()
    csv_writer.writerow(header)

    for timestamp in data:
        csv_writer.writerow(timestamp.values())
    
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
import re
import numpy as np

pd.options.display.max_rows = 5

# cleaning
df = pd.read_csv(OUT_PATH)
df['event'] = df['event'].str.replace(r'avaliable', 'available') # typo in data
df
```

### TIMES

```python
cond1 = df['event'].str.contains('Task')
cond2 = df['event'].str.contains('Ended')

task_overview_df = df[cond1 | cond2]

task_overview_df
```

```python
times = [time for time in task_overview_df['time']]

dur_times = []
i = 0
while i < len(times):
    dur_times.append(times[i + 1] - times[i])
    i += 2

between_times = []
i = 1
while i < len(times) - 1:
    between_times.append(times[i + 1] - times[i])
    i += 2    
```

```python
def avg(lst):
    return sum(lst) / len(lst)

def plot_hist(lst):
    plt.hist(lst, bins=10, edgecolor='k')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.title('Histogram')
    plt.show()

print("AVG TASK DURRATION:", avg(dur_times))
plot_hist(dur_times)

print("AVG WAIT DURRATION:", avg(between_times))
plot_hist(between_times)
```

```python
cond1 = df['event'].str.contains('Entered home screen')
cond2 = df['event'].str.contains('Task')

choices_df = df[cond1 | cond2 ]

# each job is going to be a list of events
# [
#     (time, event),
#     (time, event),
#     (time, event),
# ]
choices = []

entries = choices_df.to_records(index=False)

i = 0
while i < len(entries) - 1:
    events = []

    curr_time = entries[i][0]

    pattern = r"\"i:\(\d+,\d+\) (.*?): (\d+) waitTime, (\d+) timeLimit\""
    matches = re.findall(pattern, entries[i][1])
    for match in matches:
        # print(f"Job: {match[0]}, WaitTime: {match[1]}, TimeLimit: {match[2]}")
        events.append((curr_time + int(match[1]), match[0] + " T:" + match[2]))

    events.append((entries[i+1][0], "started " + entries[i+1][1][5:]))
    events.sort(key=lambda x: x[0])

    choices.append(events)
    i += 2

choices
```

```python
def plot_timeline(events):
    # Extract time and event descriptions
    times, descriptions = zip(*events)

    # Create a horizontal line to represent the timeline
    plt.hlines(1, min(times), max(times), linewidth=1, color='black', label="Timeline")

    # Plot event markers on the timeline with conditional colors
    for time, desc in zip(times, descriptions):
        if "started" in desc:
            plt.plot(time, 1, 'bo', markersize=3, label="Chose this task")
            plt.text(time, 1.1, desc, ha='center')
        else:
            plt.plot(time, 1, 'ro', markersize=3, label="Event")

    # Customize the plot
    plt.xlabel("Time")
    plt.yticks([])  # Remove y-axis
    plt.legend(loc='upper left')
    plt.title("Timeline")
    plt.grid(False)

    plt.show()

# CHANGE THE INDEX
plot_timeline(choices[3])
```

```python
first_chosen = 0
chosen_in_5 = 0

for choice in choices:
    before = 0

    for entry in choice:
        if "started" in entry[1]:
            if before == 1:
                first_chosen += 1
            if float(entry[0]) - float(choice[0][0]) < 5:
                chosen_in_5 += 1
        before += 1

print("Chose first:\t", first_chosen)
print("Chose in 5 s:\t", chosen_in_5)
print("Total tasks:\t", len(choices))
```

### Ending result summary

```python
earnings_df = df[df['event'].str.contains('Ended')]

total_earned = 0

for event in earnings_df["event"]:
    
    pattern = r"(\d+) earned"
    matches = re.findall(pattern, event)

    total_earned+= float( matches[0])

print("TOTAL EARNED:", total_earned)
```
